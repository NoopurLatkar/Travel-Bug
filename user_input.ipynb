```python
#This file takes in user input for the travel planner 
import datetime
def user_input_function(list_cities):
    
    flag_origin_city_check=False
    flag_destination_city_check= False
    
    list_cities_lower= list()
    for i in list_cities:
        list_cities_lower.append(i.lower())
        
    
    while(flag_origin_city_check!=True):
        
        origin_city= input("Enter the city you will be departing from: ")
        if(origin_city is None or len(origin_city)==0):
            print("Enter valid data ")
        elif((origin_city.lower().strip() not in list_cities_lower)):
            print("Please enter a valid origin city")
        elif(origin_city.lower().strip() in list_cities_lower):
            flag_origin_city_check=True
        if(flag_origin_city_check):
            break
        else:
            continue
        origin_city= input("Enter the city you will be departing from: ")
    
    while(flag_destination_city_check!=True):
    
        destination_city= input("Enter the city you want to travel to: ")
        if(destination_city is None or len(destination_city)==0):
            print("Enter valid data ")
        elif(destination_city.lower().strip() not in list_cities_lower):
            print("Please enter a valid destination city ")
        elif(origin_city.lower().strip() in list_cities_lower):
            flag_destination_city_check=True
            break
        if(flag_destination_city_check):
            break
        else:
            continue
        destination_city= input("Enter the city you want to travel to: ")
       
    trip_type= input("Is it single or a round trip? ")
    trip_type_flag= False
    while(trip_type_flag!=True):
        if(check_trip_type(trip_type)==True):
            trip_type_flag=True
            break
        else:
            print("Please enter a valid trip type")
            
        trip_type= input("Is it single or a round trip? ")
        

    flag_date= False
    if(trip_type.lower().strip()=="single"):
        while(flag_date!=True):
            start_date= input("Enter the departure date in YYYY-MM-DD: ")
            if(date_parser(start_date)):
                flag_date=True
            else:
                print("Please enter a valid date in YYYY-MM-DD :)")
            if(flag_date==True):
                break
            else:
                 continue
            start_date= input("Enter the departure date in YYYY-MM-DD: ")

    elif(trip_type.lower().strip()=="round"):
        flag_departure_date=False
        flag_arrival_date=False
        while((flag_departure_date!=True) and (flag_arrival_date!=True)):
            start_date_round= input("Enter the departure date in YYYY-MM-DD: ")
            return_date_round= input("Enter the return date in YYYY-MM-DD: ")
            if((date_parser(start_date_round))and (date_parser(return_date_round))):
                flag_departure_date=True 
                flag_arrival_date=True
                break
            else:
                print("Please enter valid dates")
            start_date_round= input("Enter the departure date in YYYY-MM-DD: ")
            return_date_round= input("Enter the return date in YYYY-MM-DD")

        
    print(origin_city)
    print(destination_city)
    print(trip_type)

def date_parser(date_str):
    try:
        return bool(datetime.datetime.strptime(date_str, '%Y-%m-%d'))
    except ValueError:
        return False
    
def check_trip_type(trip_type_str):
    if((trip_type_str.lower().strip()=="single")or (trip_type_str.lower().strip()=="round")):
        print(trip_type_str)
        return True
    else:
        return False
        
    
if __name__=="__main__":
    user_input_function(['Chicago','New York','Pittsburgh','Los Angeles'])
        
```
